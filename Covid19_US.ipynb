```python
df_daily_sum = df.groupby('Province_State').agg({'Confirmed':'sum','Deaths':'sum','Recovered':'sum'})


if 'Recovered' in df.index:
    df_daily_sum = df_daily_sum.drop(['Recovered']) # finding the province wise confirmed,recovered & deaths cases in US.

df_daily_sum
```

```python
df_daily_cases = df_daily_sum.iloc[:, [1]]
df_daily_cases
```
